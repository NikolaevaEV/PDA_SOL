```python
df = pd.read_csv("kaggle_survey_2022_responses.csv")
df.head(10)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.shape
```

```python
my_df =df[['Q2','Q3','Q4','Q5','Q8','Q9','Q11','Q16','Q23','Q24','Q25','Q29']].copy()
```

```python
my_df.columns = ['age', 'gender', 'country', 'student?', 'education', 'publications', 'years of programming', 'years of using ML', 'profession', 'industry', 'size of company', 'salary']
```

```python
my_df = my_df.drop (index= 0 )
```

```python
figure, axs = plt.subplots(2, 1, figsize=(15, 5))

axs[0].hist(my_df['gender'])
axs[0].set_xlabel('Пол')
axs[0].set_ylabel('Кол-во человек')
axs[0].set_title('Распределение по гендерному признаку')

axs[1].hist(my_df['age'])
axs[1].set_xlabel('Возраст')
axs[1].set_ylabel('Кол-во человек')
axs[1].set_title('Распределение по возрастному признаку')

plt.tight_layout()
plt.show()
```

```python
data = my_df['gender'].value_counts()
plt.figure(figsize = (8,4))

plt.pie(data, autopct='%1.0f%%')

plt.title('Распределение участников опроса по половому признаку')
plt.legend(data.index, loc='lower right', bbox_to_anchor = (1.5, 0), fontsize="8")

plt.show()
```

```python
data = my_df['student?'].value_counts()
plt.figure(figsize = (8,4))

plt.pie(data, autopct='%1.1f%%')

plt.title('Является ли участник опроса в данный момент обучающимся')
plt.legend(data.index, loc='lower right', bbox_to_anchor = (1.5, 0), fontsize="8")

plt.show()
```

```python
name=my_df.columns[0]
age=my_df.groupby(name)[name].count()

ax =age.plot(kind="bar",figsize=(8,4))
plt.ylabel('Кол-во человек')
ax.set_xlabel('Возраст')
plt.title('Распределение участников опроса по возрастному признаку')
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
for i in ax.patches:
    ax.annotate(i.get_height(),(i.get_x()+i.get_width()/2,i.get_height()+100), ha='center', va='center')
```

```python
my_df.country.value_counts(normalize=True).sort_values(ascending=False)[14::-1].plot(kind="barh")
plt.title('Распределение участников опроса по странам')
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
ax=my_df["years of programming"].value_counts().plot(kind='bar',figsize=(8,4))
plt.ylabel('Кол-во человек')
ax.set_xlabel('Кол-во лет')
plt.title('Опыт написания кода/программирования')
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
for i in ax.patches:
    ax.annotate(i.get_height(),(i.get_x()+i.get_width()/2,i.get_height()+150), ha='center', va='center')
```

```python
my_df.education.value_counts(normalize=True).sort_values(ascending=False)[14::-1].plot(kind="barh")
plt.title('Уровень образования участников опроса')
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
fig, axs = plt.subplots()
cross = pd.crosstab(my_df['years of programming'], my_df['years of using ML'], normalize='columns', margins=False)
sns.heatmap(cross, cmap="Blues", annot=True, cbar=False, linewidths=0.5)
axs.set_xlabel('Опыт использования методов ML')
axs.set_ylabel('Опыт программирования')
plt.show()
```
